```python

import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
selected_features= ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combining all the selected features
combined_feature = movies_data['genres']+' '+ movies_data['keywords'] + ' ' + movies_data['tagline'] + ' ' + movies_data['cast']+' '+ movies_data['director']
```

```python
print(combined_feature)
```

```python
#converting the text data to feature vectors
combined_feature = combined_feature.fillna('')
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vectors)
```

```python
#getting the similarity score using cosine_similarity
similarity = cosine_similarity(feature_vectors)
```

```python
#getting the movie name from the user
movie_name = input('Enter your favourite movie name: ')
```

```python
#creating the list with all the movie names given in the dataset
list_of_all_titles = movies_data['title'].tolist()
```

```python
#finding the close match for the movie name given by user
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_movie)

```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)

len(similarity_score)
```

```python
#sorting the movies based on the similarity score
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1],reverse = True)
print(sorted_similar_movies)
```

```python
#print the name of similiar movies based on the similarity score
print('Movies suggested for you: \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if(i<=30):
    print(i,'.',title_from_index)
    i+=1
```

Movie Recommendation System
